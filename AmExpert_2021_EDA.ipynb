```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df_amex' is your dataframe and 'age' is the column with age data

# Define the bins for the age groups (21-25, 26-30, ..., 51-55)
bins = list(range(20, 56, 5))
labels = ['21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55']

# Create a new column in the dataframe for the age groups
df_amex['age_group'] = pd.cut(df_amex['age'], bins=bins, labels=labels, right=False)

# Count the number of people in each age group
age_group_counts = df_amex['age_group'].value_counts().sort_index()

# Plot a bar chart for the age groups
plt.figure(figsize=(6, 4))
age_group_counts.plot(kind='bar', color='orange')

# Add labels and title
plt.title('Age Group Distribution', fontsize=12)
plt.xlabel('Age Group', fontsize=10)
plt.ylabel('Number of People', fontsize=10)

# Display the bar chart
plt.show()

# Calculate the average number of people per age group
average_count_per_age_group = age_group_counts.mean()

print("Average number of people per age group:", average_count_per_age_group)
```

```python
# Count the occurrences of each gender
gender_counts = df_amex['gender'].value_counts()

gender_percentages = (gender_counts / gender_counts.sum()) * 100

plt.figure(figsize=(5,4))
ax = gender_counts.plot(kind='bar', color=['#967969', '#967969'], edgecolor='black')

# Add percentage labels
for i, count in enumerate(gender_counts):
    percentage = gender_percentages[i]
    ax.text(i, count + 0.01 * count, f'{percentage:.1f}%', ha='center', fontsize=10)

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Keep the x labels horizontal
plt.show()
```

```python
import numpy as np

# Define logarithmic bins to handle the large range of credit limit values
bins = np.logspace(np.log10(4000), np.log10(31200000), num=30)

# Histogram for 'credit_limit' and store the counts and bin edges
plt.figure(figsize=(10, 6))
counts, bin_edges, patches = plt.hist(df_amex['credit_limit'], bins=bins, color = '#808000', edgecolor = 'black')

# Set the x-axis to a logarithmic scale
plt.xscale('log')

# Add data labels on top of each bar
for count, patch in zip(counts, patches):
    bin_x = patch.get_x() + patch.get_width() / 2
    plt.text(bin_x, count + 0.01 * count, f'{int(count):,}', ha = 'center', va = 'bottom', fontsize = 8)

# Titles and labels
plt.title('Distribution of Credit Limit (Including Outlier)')
plt.xlabel('Credit Limit (Log Scale)')
plt.ylabel('Frequency')

# Add thousands separator to the y-axis
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.show()
```

**Independent Variable Relationship with credit_card_default**

**Categorical Variables**: gender, owns_car, owns_house, migrant_worker

**Numerical Variables**: age, net_yearly_income, credit_limit

```python
# Calculate the percentage of defaults and no defaults by gender
gender_default = df_amex.groupby('gender')['credit_card_default'].value_counts(normalize=True).unstack() * 100

# Plot percentage as a bar chart
plt.figure(figsize=(8, 5))
ax = gender_default.plot(kind='bar', stacked=True, color=['#967969', '#ffcccb'], edgecolor='black', figsize=(8, 6))

# Add data labels to the bars for percentage
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x = p.get_x()
    y = p.get_y()
    ax.annotate(f'{height:.2f}%', (x + width/2, y + height/2),
                ha='center', va='center', fontsize=10, color='black')

# Titles and labels
plt.title('Percentage of Credit Card Default by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```
